import dependencies

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

open an image

```python
os.chdir(os.getcwd()+'/images')
```

```python
def get_filenames():
    input_path = []
    for image_path in os.listdir():
        input_path.append(os.path.join(os.getcwd(), image_path))
    return(input_path)
input_paths = get_filenames()
```

```python
spirochete = cv2.imread(input_paths[0], cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(spirochete, cv2.COLOR_BGR2RGB))
plt.show()
```

Demo a flipped image

```python
flipped = cv2.flip(spirochete, 1)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.show()
```

Demo a clockwise rotated image

```python
image = cv2.rotate(spirochete, cv2.ROTATE_90_CLOCKWISE)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

Make mirror and clockwise roated images

```python
input_paths = [f for f in input_paths if (f.endswith(".jpg"))]
```

```python
str(os.path.split(input_paths[0])[1]).replace('.jpg', '')
```

```python
index = 0

for path in input_paths:
    index = index+1
    img = cv2.imread(path, cv2.COLOR_BGR2RGB)
    newimg = cv2.flip(img, 1)
    filename = str(os.path.split(path)[1]).replace('.jpg', '') + 'mirror.jpg'
    cv2.imwrite(filename, newimg)
```

```python
index = 0

for path in input_paths:
    index = index+1
    img = cv2.imread(path, cv2.COLOR_BGR2RGB)
    newimg = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    filename = str(os.path.split(path)[1]).replace('.jpg', '') + 'clock.jpg'
    cv2.imwrite(filename, newimg)
```
